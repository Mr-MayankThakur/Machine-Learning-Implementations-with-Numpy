# matplotlib tutorial

```
"""
Using numpy to generate random data
"""

import numpy as np

x = np.linspace(0, 10, 25)
y = x * x + 2

print(x)
print(y)
```

```
# importing matplotlib here

from matplotlib import pylab
```

```
# now drawing graph for the data

pylab.plot(x,y, 'r') # r is red
```

```
# drawing subgraphs

pylab.subplot(1,2,1) # (rows, columns, indexes)
pylab.plot(x,y,'r--') # red and dash line
pylab.subplot(1,2,2)
pylab.plot(y,x,'g--')
```

```
# now we are importing pyplot module

from matplotlib import pyplot as plt
```

```
# repeating earlier process with pyplot

fig = plt.figure()

axes = fig.add_axes([0.5,0.1,0.8,0.8]) #control the left, right, width, height of the canvas(from 0 to i)

axes.plot(x,y,'r')
```

```
# again we draw subgraphs

fig, axes = plt.subplots(nrows= 1, ncols=2) # submap is of 1 row and  cols
for ax in axes:
  ax.plot(x,y,'r')
```

```
# we can also draw a picture or graph inside another graph

fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # big axis
axes2 = fig.add_axes([0.2,0.5,0.4,0.3]) # small canvas

axes1.plot(x,y,'r')
axes2.plot(y,x,'g--')
```

```
ax.legend(['label1','label2'])
fig, axes = plt.subplots(dpi=90) # we can set dpi for figure also
axes.set_xlabel('x-label')
axes.set_ylabel('y-label')

axes.set_title("title")

axes.plot(x, x**2)
axes.plot(x,x**3)

axes.legend(["y=x**3", "y=x**3"],loc=2)
```

```
# in matplotlib you can set other properties such as line color, transparency, and more

fig, axes = plt.subplots(dpi=150)

axes.plot(x, x**2, color='red', alpha = 0.3)
axes.plot(x,x+2, color="#1155dd", alpha=0.5)
axes.plot(x, x+3, color="green")
```

```
# for line widths

fig, ax = plt.subplots(dpi=100)

ax.plot(x,x+1, color='blue', linewidth=0.25)
ax.plot(x,x+2, color='blue', linewidth=1)
ax.plot(x,x+3, color='blue', linewidth=1.5)
ax.plot(x,x+4, color='blue', linewidth=2)
```

```
# for line styles

fig, ax = plt.subplots(dpi=100)

ax.plot(x,x+1, color='blue', lw=2, linestyle='-')
ax.plot(x,x+2, color='blue', lw=2, linestyle='-.')
ax.plot(x,x+3, color='blue', lw=2, linestyle=':')
```

```
# markers -> o, s, +, 1

fig, ax = plt.subplots(dpi=100)

ax.plot(x,x+1, color='blue', marker='o')
ax.plot(x,x+2, color='blue', marker='+')
ax.plot(x,x+3, color='blue', marker='s')
ax.plot(x,x+4, color='blue', marker='1')
ax.plot(x,x+5, color='blue', marker='o', markersize=8, markerfacecolor='red')
```

```
# set canvas grid and axis size

fig, Axes = plt.subplots(1,2,figsize=(10,5))

Axes[0].plot(x,x**2, x,x**3, lw=2)
Axes[0].grid()

Axes[1].plot(x,x**2, x,x**3, lw=2)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([0,5])
Axes[1].grid()
```

# Other 2D graphs

```
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1, 4, figsize=(16,5))
axes[0].set_title("scatter")
axes[0].scatter(x,x + 0.25*np.random.rand(len(x)))

axes[1].set_title("step")
axes[1].step(n,n**2, lw=2)

axes[2].set_title("bar")
axes[2].bar(n,n**2, align='center', alpha=0.5)

axes[3].set_title("fill-between")
axes[3].fill_between(x,x**2,x**3, color='green', alpha=0.5)
```

```
# draw a radar chart

fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0,0.0,1,1],polar=True)

t = np.linspace(0,2*np.pi,100)
ax.plot(t,t,color='blue', lw=3)
```

```
# Draw a histogram

n = np.random.randn(100000)

fig, axes = plt.subplots(1,2,figsize=(12,4))
axes[0].set_title("Default Histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative Detailed Histogram")
axes[1].hist(n, cumulative=True, bins=50)
```

```
# Draw contour maps

# generating data to plot
import matplotlib
import numpy as np
import matplotlib.cm as cm # color maps
import matplotlib.pyplot as plt

delta = 0.025
x=np.arange(-3.0, 3.0, delta)
y=np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x,y)  
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-1)**2 - (Y-1)**2)
Z = (Z1 - Z2) * 2

# plotting the graph
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS, inline=1, fontsize=10)
```

# 3D graphs

```
from mpl_toolkits.mplot3d.axes3d import Axes3D

fig = plt.figure(figsize=(14,6))

# specify 3D graphics to draw, with projection = 3d
ax = fig.add_subplot(1,1,1,projection='3d') 

ax.plot_surface(X,Y,Z, rstride=4,cstride=4, linewidth=0)
```
